```python
# 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_inline.backend_inline import set_matplotlib_formats

# 设置输出格式为SVG
set_matplotlib_formats("svg")


# 定义各种激活函数
def Logistic(x):
    return 1 / (1 + np.exp(-x))


def Tanh(x):
    return np.tanh(x)


def ReLU(x):
    return np.maximum(0, x)


def leaky_ReLU(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)


def softplus(x):
    return np.log(1 + np.exp(x))


# 生成横坐标
x = np.linspace(-2, 2, 1000)
# 计算各种激活函数在以上横坐标上的值
y_logistic = Logistic(x)
y_tanh = Tanh(x)
y_relu = ReLU(x)
y_leaky_relu = leaky_ReLU(x)
y_softplus = softplus(x)

# 创建绘图画布和坐标轴
fig, ax = plt.subplots(figsize=(9, 6))

# 隐藏画布上右侧和顶部的边框
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# 添加水平和垂直实线
ax.axhline(y=0, color="k")
ax.axvline(x=0, color="k")

# 将坐标轴底部和左侧的边框移到0的位置
ax.spines["bottom"].set_position("zero")
ax.spines["left"].set_position("zero")

# 绘制各种激活函数曲线
plt.plot(x, y_logistic, label="Logistic")
plt.plot(x, y_tanh, label="Tanh")
plt.plot(x, y_relu, label="ReLU", linewidth=4)
plt.plot(x, y_leaky_relu, label="Leaky ReLU")
plt.plot(x, y_softplus, label="Softplus")

# 添加 y=1 和 y=-1 的水平虚线
plt.axhline(y=1, ls="--", c="gray")
plt.axhline(y=-1, ls="--", c="gray")

# 添加图例
plt.legend()

# 添加标题
plt.title("Activation Functions")

# 将绘制结果保存为SVG格式，并命名为"activation_functions"
plt.savefig("activation_functions.svg", format="svg", bbox_inches="tight")
```
